```python
txt = df['review']
number_of_words = []
for i in txt :
  a = i.split()
  number_of_words.append(len(a))

df['number_of_words'] = number_of_words
```

```python
df['sentiment']=pd.get_dummies(df['sentiment'])['positive']
df.head()
```
